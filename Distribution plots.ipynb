4.rugplot

```python
import seaborn as sns
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

# Hist plot / displot

```python
df=sns.load_dataset('tips')
df
```

```python
df['size'].unique()  #That means only 'total_bills and tips' hi quantitative hai
```

```python
sns.displot(df['total_bill'],bins=20,kde=True)  

#default me 20 bins hi hote hai
#histogram ka hi use karke hum kernal density estimator curve bante hai 
```

```python
sns.histplot(df['tip'],bins=20,kde=True)    #displot and histplot dono sam ehi hai 
```

```python
plt.subplot(1,2,1)
sns.displot(df['total_bill'],bins=20,kde=True)

plt.subplot(1,2,2)
sns.histplot(df['tip'],kde=True)


#Doubt hai
```

# Joint Plot >> 2 Data ko sath me compare karna hai

Shift + tab click karke sare parameters pata chal jate hai 

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='resid')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=df,kind='hex')
```

# Pair Plot

```python
sns.pairplot(df,hue="sex")
```

```python
sns.pairplot(df,hue="size",palette='rainbow')
```

# Rugplot 

```python
sns.rugplot(df)
```
