```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Matplotlib basics

```python
x=np.linspace(0,5,11)
print(x)
y=x**3
print(y)
```

```python
plt.plot(x,y)
plt.title("Graph of x^3")
plt.xlabel("Number")
plt.ylabel("Cubes")
```

# Subplots in matplotlib

```python
plt.subplot(2,2,1)     #(canvas size(breadth),canvas size(height),selected portion)
plt.plot(y,x)          #jis subset kpar plot karna uske just neeche code type karo


plt.subplot(2,2,4)
plt.plot(x,y)


plt.subplot(2,2,2)
plt.plot(x**2,y)


plt.subplot(2,2,3)
plt.plot(x,y**2)
```

# Object Oriented Way >> For selecting the whole canvas 

```python
fig=plt.figure(figsize=(8,6),dpi=100)

axis1= fig.add_axes([.1,.1,.8,.8])  #([left,bottom,width,height])
axis1.plot(x,y)

axis2= fig.add_axes([.3,.3,.5,.4])
axis2.plot(x,y)
```

```python
fig = plt.figure()
ax=fig.add_axes([.1,.1,.8,.8])
ax.plot(x,y)
plt.savefig("basicplot.png")
```
