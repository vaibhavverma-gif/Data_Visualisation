```python
import seaborn as sns
import matplotlib.pyplot as plt
```

# lmplot()
regression plot is used in machine learning so we will not fully discuss what, how and where it is used but we will see how it is created and what are the values we can put inside it

## Regression Plot

* x axis : JIske through prediction hota hai 
* y axis : Prediction

```python
tips=sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)  #similar to joint plot
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='coolwarm',markers=['o','v'],
           scatter_kws={'s':60})
```
